```python
from sklearn.preprocessing import LabelEncoder,OrdinalEncoder
la=LabelEncoder()
OR=OrdinalEncoder()
property['homeType']=la.fit_transform(property[['homeType']])
property['homeStatus']=la.fit_transform(property[['homeStatus']])
property['state']=la.fit_transform(property[['state']])
property['city']=la.fit_transform(property[['city']])
#property['county']=la.fit_transform(property[['county']])
property
```

```python
property['datePosted']=pd.to_datetime(property['datePosted'],format='mixed')
dates=property['datePosted']
property.drop('datePosted',axis=1,inplace=True)
property.drop('lastUpdated',axis=1,inplace=True)
property.drop('zipcode',axis=1,inplace=True)
property.drop('rentZestimate',axis=1,inplace=True)
property.drop('pageViewCount',axis=1,inplace=True)
property.drop('favoriteCount',axis=1,inplace=True)
#property.drop('yearBuilt',axis=1,inplace=True)
property
```

```python
property['bedrooms'].value_counts()
property['bedrooms'].astype(int)
property['bedrooms'].astype(int)
property=property[(property['bathrooms']<18) & (property['bedrooms']<18)]
print(property['bathrooms'].value_counts())
```

```python
property=property[property['price']>0]
property['price'].value_counts()
```

```python
y=property['price']
property.drop('price',axis=1,inplace=True)
y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error,r2_score
from sklearn.linear_model import LinearRegression,Ridge,Lasso
from sklearn.ensemble import RandomForestRegressor,AdaBoostRegressor
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor
X_train,X_valid,y_train,y_valid=train_test_split(property,y,train_size=0.8,test_size=0.2,random_state=42)
model1=LGBMRegressor(n_estimators=500,random_state=42)
model2=XGBRegressor(n_estimators=500,random_state=42)
model3=RandomForestRegressor(n_estimators=500,random_state=42)
```

```python
model1.fit(X_train,y_train)
model2.fit(X_train,y_train)
model3.fit(X_train,y_train)
```

```python
model1.fit(property,y)
model2.fit(property,y)
model3.fit(property,y)
```

```python
pred1=model1.predict(X_valid)
print(pred1)
print(mean_absolute_error(y_valid,pred1))
print(r2_score(y_valid,pred1))
```

```python
pred2=model2.predict(X_valid)
print(pred2)
print(mean_absolute_error(y_valid,pred2))
print(r2_score(y_valid,pred2))
```

```python
pred3=model3.predict(X_valid)
print(pred3)
print(mean_absolute_error(y_valid,pred3))
print(r2_score(y_valid,pred3))
```
